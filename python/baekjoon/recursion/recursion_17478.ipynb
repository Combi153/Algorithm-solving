recursive 17478 재귀함수 챗봇

JH 교수님은 재귀함수가 무엇인지 물어보는 학생들을 위한 자동 응답 챗봇을 만들었다.
- 교수님이 출력을 원하는 재귀 횟수 N을 입력한다.
- 챗봇의 응답을 출력한다.

```python
n=int(input())
a=0

def chatbot(n,a):
    line='_'*a
    print('{}"재귀함수가 뭔가요?"'.format(line))
    if n==0:
        print('{}"재귀함수는 자기 자신을 호출하는 함수라네"'.format(line))
        print('{}라고 답변하였지.'.format(line))
        return

    else:    
        print('{}"잘 들어보게. 옛날옛날 한 산 꼭대기에 이세상 모든 지식을 통달한 선인이 있었어.'.format(line))
        print('{}마을 사람들은 모두 그 선인에게 수많은 질문을 했고, 모두 지혜롭게 대답해 주었지.'.format(line))
        print('{}그의 답은 대부분 옳았다고 하네. 그런데 어느 날, 그 선인에게 한 선비가 찾아와서 물었어."'.format(line))

    chatbot(n-1,a+4)

    print('{}라고 답변하였지.'.format(line))

print("어느 한 컴퓨터공학과 학생이 유명한 교수님을 찾아가 물었다.")
chatbot(n,a)
```

다른 사람이 쓴 코드 
- 변수의 유효 범위에 대한 오해

```python
def recursive(m):
    print("_" * (4 * (n - m)) + '"재귀함수가 뭔가요?"')

    if not m:
        print("_" * (4 * (n - m)) + '"재귀함수는 자기 자신을 호출하는 함수라네"')
        print("_" * (4 * (n - m)) + "라고 답변하였지.")
        return

    print("_" * (4 * (n - m)) + '"잘 들어보게. 옛날옛날 한 산 꼭대기에 이세상 모든 지식을 통달한 선인이 있었어.')
    print("_" * (4 * (n - m)) + "마을 사람들은 모두 그 선인에게 수많은 질문을 했고, 모두 지혜롭게 대답해 주었지.")
    print("_" * (4 * (n - m)) + '그의 답은 대부분 옳았다고 하네. 그런데 어느 날, 그 선인에게 한 선비가 찾아와서 물었어."')
    recursive(m - 1)
    print("_" * (4 * (n - m)) + "라고 답변하였지.")


n = int(input())
print("어느 한 컴퓨터공학과 학생이 유명한 교수님을 찾아가 물었다.")
recursive(n)
```

변수의 유효 범위
- 함수 내부에서 선언된 변수를 지역변수라 하고, 함수 외부에서 선언된 변수를 전역 변수라고 한다.
- 함수 내부에서 선언된 지역 변수는 해당 함수 내부에서만 사용할 수 있다. 함수 밖에서는 사용할 수 없다.
- 함수 내부에서 선언된 지역변수를 함수 밖에서 활용하고 싶다면, global을 통해 전역 변수로 선언해야 한다.

- 단, 함수 밖에서 선언된 전역 변수를 함수 내부에서 사용하는 것은 가능하다.
